```python
import requests
import json

def upload_pdf_and_get_text(pdf_path):
    files = {"file": open(pdf_path, "rb")}
    response = requests.post("http://localhost:8000/pdf-to-text/", files=files)

    if response.status_code == 200:
        print("✅ PDF uploaded successfully!")
        return response.json().get('text', '')
    else:
        print("❌ Failed to upload PDF. Status code:", response.status_code)
        print(response.json())
        return None

def generate_quiz_with_openrouter(pdf_text, openrouter_api_key):
    prompt = f"Generate a quiz with 10 questions from the following content:\n{pdf_text[:5000]}"

    headers = {
        "Authorization": f"Bearer {openrouter_api_key}",
        "Content-Type": "application/json",
    }

    data = json.dumps({
        "model": "deepseek/deepseek-chat-v3-0324:free",
        "messages": [
            {"role": "user", "content": prompt}
        ],
    })

    response = requests.post("https://openrouter.ai/api/v1/chat/completions", headers=headers, data=data)

    if response.status_code == 200:
        print("✅ Quiz generated successfully!")
        print(response.json()['choices'][0]['message']['content'])
    else:
        print("❌ Failed to generate quiz. Status code:", response.status_code)
        print(response.json())

def main():
    pdf_path = "sample.pdf"  # Replace with your PDF path
    openrouter_api_key = "<OPENROUTER_API_KEY>"  # Replace with your OpenRouter API key

    # Step 1: Upload PDF and extract text
    pdf_text = upload_pdf_and_get_text(pdf_path)

    if pdf_text:
        # Step 2: Generate quiz from the extracted text
        generate_quiz_with_openrouter(pdf_text, openrouter_api_key)

if __name__ == "__main__":
    main()
```
